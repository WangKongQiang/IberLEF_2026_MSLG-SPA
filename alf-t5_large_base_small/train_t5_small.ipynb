MSLG_to_SPA

```python
from alf_t5 import ALFT5Translator

# Example language data
# .txt, .csv
# data = open("my_data.txt", "r").read()
data = open("MSLG_to_SPA_data.txt", "r").read()
# Create and train translator
translator = ALFT5Translator(
    model_name="models--google-t5--t5-small",
    use_peft=True,
    batch_size=4,
    num_epochs=20,
    output_dir="MSLG_to_SPA_alf_t5_small_translator"
)

# Train the model
translator.train(
    data_string=data,
    test_size=0.2,
    # augment=True
     augment=False   # ✅ 关掉
)
```

```python
translation = translator.translate("PEZ COLOR AZUL MI FAVORITO", direction="c2e")
translation
```

```python
translation = translator.translate("Los peces azules son mis favoritos.", direction="e2c")
translation
```

MSLG2SPA

读取代码：

```python
test_data = []

with open("MSLG2SPA_test.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()[1:]  # 跳过表头
    
    for line in lines:
        parts = line.strip().split("\t")
        if len(parts) == 2:
            idx, text = parts
            test_data.append((idx, text))
```

✅ 批量预测

```python
results = []

for idx, text in test_data:
    try:
        translation = translator.translate(text, direction="c2e")
    except Exception as e:
        print(f"Error on {idx}: {e}")
        translation = ""
    
    results.append((idx, translation))
```

✅  保存结果（推荐）

```python
with open("MSLG2SPA_predictions.txt", "w", encoding="utf-8") as f:
    f.write("ID\tPREDICTION\n")
    for idx, pred in results:
        f.write(f"{idx}\t{pred}\n")
```
