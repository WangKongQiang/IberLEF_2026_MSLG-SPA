MSLG2SPA

读取代码：

```python
test_data = []

with open("MSLG2SPA_test.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()[1:]  # 跳过表头
    
    for line in lines:
        parts = line.strip().split("\t")
        if len(parts) == 2:
            idx, text = parts
            test_data.append((idx, text))
```

✅ 4. 批量预测

```python
results = []

for idx, text in test_data:
    try:
        translation = translator.translate(text, direction="c2e")
    except Exception as e:
        print(f"Error on {idx}: {e}")
        translation = ""
    
    results.append((idx, translation))
```

✅ 5. 保存结果（推荐）

```python
with open("MSLG2SPA_predictions.txt", "w", encoding="utf-8") as f:
    f.write("ID\tPREDICTION\n")
    for idx, pred in results:
        f.write(f"{idx}\t{pred}\n")
```

🚀 强烈建议（更干净的方案）

你已经训练好了模型，其实最正确的方法是👇

直接加载 LoRA：

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
from peft import PeftModel

tokenizer = T5Tokenizer.from_pretrained("models--google-t5--t5-base")

model = T5ForConditionalGeneration.from_pretrained("models--google-t5--t5-base")
model = PeftModel.from_pretrained(model, "MSLG_to_SPA_coal_t5_translator/final_model")

model.eval()
```

然后：

```python
inputs = tokenizer("PEZ COLOR AZUL MI FAVORITO", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
