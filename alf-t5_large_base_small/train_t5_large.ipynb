## Prepare for training

Import **ALFT5Translator**, load the dataset & check it out. Finally, load **T5** on **ALF** translator instance.

## Train ALF-T5

```python
translator.train(
    data_string=data,
    test_size=0.2,
    augment=True
)
```

### MSLG_to_SPA

Training ALF on T5-Large

Prepare for training

Import ALFT5Translator, load the dataset & check it out. Finally, load T5 on ALF translator instance.

```python
from alf_t5 import ALFT5Translator
```

```python
data = open("MSLG_to_SPA_data.txt", "r").read()
data
```

```python
translator = ALFT5Translator(
    model_name="models--google-t5--t5-large",
    use_peft=True,
    use_fp16=False, # Needs to be set to False, see the README for more information.
    batch_size=4,
    num_epochs=20, # It's recommended to set the number of epochs to this number, but you can change if you want.
    output_dir="MSLG_to_SPA_alf_t5_translator" # You can change the output folder if you want.
)
```

Train ALF-T5

```python
translator.train(
    data_string=data,
    test_size=0.2,
    # augment=True
    augment=False   # ✅ 关掉
)
```

Tests

MSLG-->Conlang to SPA-->English

```python
translation = translator.translate("...", direction="c2e")
translation
```

```python
translation = translator.translate("PEZ COLOR AZUL MI FAVORITO", direction="c2e")
translation
```

SPA-->English to MSLG-->Conlang

```python
translation = translator.translate("...", direction="e2c")
translation
```

```python
translation = translator.translate("Los peces azules son mis favoritos.", direction="e2c")
translation
```

MSLG2SPA

读取代码：

```python
test_data = []

with open("MSLG2SPA_test.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()[1:]  # 跳过表头
    
    for line in lines:
        parts = line.strip().split("\t")
        if len(parts) == 2:
            idx, text = parts
            test_data.append((idx, text))
```

✅ 批量预测

```python
results = []

for idx, text in test_data:
    try:
        translation = translator.translate(text, direction="c2e")
    except Exception as e:
        print(f"Error on {idx}: {e}")
        translation = ""
    
    results.append((idx, translation))
```

✅ 保存结果（推荐）

```python
with open("MSLG2SPA_predictions.txt", "w", encoding="utf-8") as f:
    f.write("ID\tPREDICTION\n")
    for idx, pred in results:
        f.write(f"{idx}\t{pred}\n")
```
