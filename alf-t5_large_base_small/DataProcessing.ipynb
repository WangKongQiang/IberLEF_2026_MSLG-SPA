给你一段完整的 Python 脚本，可以把 MSLG_SPA_train.txt 转换成你需要的格式（Source_language_sentence|Target_language_sentence\n），并输出为 MSLG_to_SPA_data.txt。

我按照常见的 MSLG SPA 数据格式（每两行一对：源句 + 目标句）来写，如果你的格式稍有不同我也可以帮你改。

```python
input_file = "MSLG_SPA_train.txt"
output_file = "MSLG_to_SPA_data.txt"

with open(input_file, "r", encoding="utf-8") as f:
    lines = f.readlines()

pairs = []

# 跳过表头（第一行）
for line in lines[1:]:
    parts = line.strip().split("\t")
    
    # 确保格式正确（3列：ID, MSLG, SPA）
    if len(parts) >= 3:
        source = parts[1].strip()  # MSLG
        target = parts[2].strip()  # SPA
        pairs.append(f"{source}|{target}\n")

# 写入新文件
with open(output_file, "w", encoding="utf-8") as f:
    f.writelines(pairs)

print(f"转换完成，共生成 {len(pairs)} 条数据，保存为 {output_file}")
```
